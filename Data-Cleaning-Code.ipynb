```python
import pandas as pd

data = {
    "Name": ["Ali","Sara","Ali","Ahmed","Fatima"],
    "Age": [23,None,23,30,27],
    "City": ["Hyderabad","Bangalore","Hyderabad","hyderabad","Bengaluru"],
    "Email": ["ali@gmail.com","sara@gmail.com","ali@gmail.com","ahmed@gmail.com",None]
}

df = pd.DataFrame(data)

df
```

```python
df.to_csv("raw_data.csv", index=False)
```

```python
df = df.drop_duplicates().copy()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
```

```python
df["Email"] = df["Email"].fillna("Unknown")
```

```python
df["City"] = df["City"].replace({
    "hyderabad": "Hyderabad",
    "Bangalore": "Bengaluru"
})
```
